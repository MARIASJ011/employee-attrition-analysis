```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:\\Users\\maria\\Downloads\\Dataset-7.csv")

# Display basic information
df.info()

# Display summary statistics
print(df.describe())

#display first few rows
print(df.head())
```

# 2. Problem Statement and Objectives

-Problem Statement:

-Employee attrition affects workforce stability and increases operational costs. 
Analyzing employee data can help organizations identify the key drivers of attrition and take proactive measures to improve employee retention.

-Objectives:

-Identify key factors influencing employee attrition.

# 3. Data Pre-processing Techniques with Justifications

Handling Missing Values:

-Missing values can lead to inaccurate analysis. We check for missing values and drop records if necessary.

Handling Duplicates:

-Duplicate records can bias analysis, so we remove them to ensure data integrity.




```python
# Handling missing values
df = df.dropna()

# Removing duplicate records
df = df.drop_duplicates()
```

# 4. EDA and Visualizations:

- Exploratory data analysis and visualizations help better understand data and provide valuable insights for better decision making.
- Here we have used histograms,barcharts,pie charts etc for data visualization.



```python
#Histogram: Distribution of Salary and Age

plt.figure(figsize=(12, 5))
sns.histplot(df['MonthlyIncome'], bins=30, kde=True, color='blue')
plt.title("Salary Distribution")
plt.show()


```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df['Age'], bins=30, kde=True, color='orange')
plt.title("Age Distribution")
plt.show()
```

```python
# Box Plot: Detecting Outliers in Salary

plt.figure(figsize=(10,5))
sns.boxplot(x=df['MonthlyIncome'])
plt.title
```

```python

# Pie chart to show proportion of employees who left vs stayed
attrition_counts = df['Attrition'].value_counts()
labels = ['Stayed', 'Left']
colors = ['grey', 'black']

plt.figure(figsize=(6, 6))
plt.pie(attrition_counts, labels=labels, colors=colors)
plt.title("Attrition Rate Distribution")
plt.show()
```

```python
#CORRELATION HEATMAP: Selecting only numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
# Bar Chart: Attrition by Job Role

plt.figure(figsize=(12,6))
sns.countplot(y=df['JobRole'], hue=df['Attrition'])
plt.title("Attrition by Job Role")
plt.show()
```

```python

# Analyzing employees with long gaps in promotion
plt.figure(figsize=(10,5))
sns.histplot(df['YearsSinceLastPromotion'], bins=30,kde=True, color='green')
plt.title("Years Since Last Promotion Distribution")
plt.show()
```

# 5. Business Recommendations

1. Targeted Retention Programs: 

Employees who frequently travel for work show a higher attrition rate. Companies should introduce work-life balance policies to retain these employees.

2. Salary Growth Strategy: 

Employees at higher job levels earn more, but salary increases do not always correlate with lower attrition. More transparent career progression plans can help retain employees.

3. Promotion and Career Development: 

Employees with longer gaps since their last promotion tend to leave. Companies should introduce structured career growth opportunities.


# 6. Conclusion

The exploratory data analysis provided valuable insights into employee attrition patterns.

Key findings include:

-Frequent business travelers have a higher attrition rate.

-Salary does not always influence retention; career growth opportunities are essential.

-Employees with long gaps in promotions are more likely to leave.

-By implementing targeted retention policies and career development initiatives, organizations can reduce attrition and improve workforce stability.
